## Задание 2: Выберите топ 3 страны по количеству заказчиков, выведите их названия и количество записей

```python
pd.read_sql(
    """
    SELECT Country, count(CustomerID) AS Count_customers  
    FROM Customers  
    GROUP BY Country  
    ORDER BY count(CustomerID) DESC  
    LIMIT 3
    """, con)
```

## Задание 3: Выберите перевозчика, который отправил 10-й по времени заказ, выведите его название, и дату отправления

```python
pd.read_sql(
    """
    SELECT ShipperName, OrderDate  
    FROM (SELECT ShipperName, OrderDate, OrderID  
        FROM Orders AS o INNER JOIN  
            Shippers AS s ON o.ShipperID = s.ShipperID  
        ORDER BY OrderDate, OrderID  
        LIMIT 10) AS top_10  
    ORDER BY OrderDate DESC, OrderID DESC  
    LIMIT 1
    """, con, parse_dates=['OrderDate'])
```

## Задание 4: Выберите самый дорогой заказ, выведите список товаров с их ценами

```python
pd.read_sql(
    """
    SELECT ProductName, Price  
    FROM (SELECT OrderID, SUM(Product_cost) AS Order_cost  
        FROM (SELECT OrderID, Price*Quantity AS Product_cost  
                FROM OrderDetails AS od INNER JOIN  
                    Products AS pr ON od.ProductID = pr.ProductID) AS Order_info  
        GROUP BY OrderID  
        ORDER BY SUM(Product_cost) DESC  
        LIMIT 1) AS top_order INNER JOIN  
        OrderDetails ON top_order.OrderID = OrderDetails.OrderID INNER JOIN  
        Products ON OrderDetails.ProductID = Products.ProductID 
    """, con)
```

добавлено количество товара в лоте для удобства восприятия

```python
pd.read_sql(
    """
    SELECT ProductName, Unit, Price  
    FROM (SELECT OrderID, SUM(Product_cost) AS Order_cost  
        FROM (SELECT OrderID, Price*Quantity AS Product_cost  
                FROM OrderDetails AS od INNER JOIN  
                    Products AS pr ON od.ProductID = pr.ProductID) AS Order_info  
        GROUP BY OrderID  
        ORDER BY SUM(Product_cost) DESC  
        LIMIT 1) AS top_order INNER JOIN  
        OrderDetails ON top_order.OrderID = OrderDetails.OrderID INNER JOIN  
        Products ON OrderDetails.ProductID = Products.ProductID 
    """, con)
```

## Задание 5: Какой товар больше всего заказывали по количеству единиц товара, выведите его название и количество единиц в каждом из заказов

```python
pd.read_sql(
    """
    SELECT (SELECT ProductName FROM Products WHERE Products.ProductID = Product_top.ProductID) AS ProductName, Quantity  
	FROM (SELECT ProductID, SUM(Quantity) AS Quantity_all  
		FROM OrderDetails  
		GROUP BY ProductID  
		ORDER BY Quantity_all DESC  
		LIMIT 1) AS Product_top INNER JOIN  
		OrderDetails ON Product_top.ProductID = OrderDetails.ProductID 
    """, con)
```

добавлен OrderID для удобства восприятия

```python
pd.read_sql(
    """
    SELECT (SELECT ProductName FROM Products WHERE Products.ProductID = Product_top.ProductID) AS ProductName, OrderID, Quantity  
	FROM (SELECT ProductID, SUM(Quantity) AS Quantity_all  
		FROM OrderDetails  
		GROUP BY ProductID  
		ORDER BY Quantity_all DESC  
		LIMIT 1) AS Product_top INNER JOIN  
		OrderDetails ON Product_top.ProductID = OrderDetails.ProductID 
    """, con)
```

## Задание 6: Выведите топ 5 поставщиков по количеству заказов, выведите их названия, страну, контактное лицо и телефон

```python
pd.read_sql(
    """
    SELECT SupplierName, Country, ContactName, Phone, Count_order  
      FROM (SELECT SupplierID, Count(OrderID) AS Count_order  
            FROM (SELECT DISTINCT sup.SupplierID AS SupplierID, OrderID  
                  FROM Suppliers AS sup INNER JOIN  
                  Products AS pr ON sup.SupplierID = pr.SupplierID INNER JOIN  
                  OrderDetails AS od ON pr.ProductID = od.ProductID) AS Supplier_order  
            GROUP BY SupplierID  
            ORDER BY Count_order DESC  
            LIMIT 5) AS Supplier_top INNER JOIN Suppliers  
            ON Supplier_top.SupplierID = Suppliers.SupplierID
    """, con)
```

добавлено количество заказов по поставщикам (если поставщик участует в заказе по 2 продуктам - он учитывается 1 раз)

```python
pd.read_sql(
    """
    SELECT SupplierName, Country, ContactName, Phone, Count_order  
      FROM (SELECT SupplierID, Count(OrderID) AS Count_order  
            FROM (SELECT DISTINCT sup.SupplierID AS SupplierID, OrderID  
                  FROM Suppliers AS sup INNER JOIN  
                  Products AS pr ON sup.SupplierID = pr.SupplierID INNER JOIN  
                  OrderDetails AS od ON pr.ProductID = od.ProductID) AS Supplier_order  
            GROUP BY SupplierID  
            ORDER BY Count_order DESC  
            LIMIT 5) AS Supplier_top INNER JOIN Suppliers  
            ON Supplier_top.SupplierID = Suppliers.SupplierID
    """, con)
```

## Задание 7: Какую категорию товаров заказывали больше всего по стоимости в Бразилии, выведите страну, название категории и сумму

```python
pd.read_sql(
    """
    SELECT Country, (SELECT CategoryName FROM Categories AS ctg WHERE ctg.CategoryID = Brazil_orders.CategoryID) AS CategoryName,  
           SUM(Prod_cost) AS Category_sum  
      FROM (SELECT country, CategoryID, Quantity*Price AS Prod_cost  
            FROM Customers AS cs INNER JOIN  
                  Orders AS ors ON cs.CustomerID = ors.CustomerID INNER JOIN  
            OrderDetails AS ordt ON ors.OrderID = ordt.OrderID INNER JOIN  
            Products AS pr ON ordt.ProductID = pr.ProductID  
            WHERE Country = "Brazil") AS Brazil_orders  
      GROUP BY Country, CategoryID  
      ORDER BY Category_sum DESC  
      LIMIT 1
    """, con)
```

## Задание 8: Какая разница в стоимости между самым дорогим и самым дешевым заказом из США

```python
pd.read_sql(
    """
    SELECT (MAX(Order_cost) - MIN(Order_cost)) AS Difference_MAX_MIN_orders_USA  
    FROM (SELECT ors.OrderID, SUM(Quantity*Price) AS Order_cost  
        FROM Customers AS cs INNER JOIN  
            Orders AS ors ON cs.CustomerID = ors.CustomerID INNER JOIN  
            OrderDetails AS ordt ON ors.OrderID = ordt.OrderID INNER JOIN  
            Products AS pr ON ordt.ProductID = pr.ProductID  
        WHERE Country = "USA"
        GROUP BY ors.OrderID) AS Orders_USA
    """, con)
```

## Задание 9: Выведите количество заказов у каждого их трех самых молодых сотрудников, а также имя и фамилию во второй колонке

```python
pd.read_sql(
    """
    SELECT Count(OrderID) AS Count_Orders, Name  
    FROM (SELECT EmployeeID, (FirstName || " " || LastName) AS Name  
        FROM Employees  
        ORDER BY BirthDate DESC  
        LIMIT 3) AS Youngests INNER JOIN  
        Orders ON Youngests.EmployeeID = Orders.EmployeeID  
    GROUP BY Name
    """, con, parse_dates=['BirthDate'])
```

## Задание 10: Сколько банок крабового мяса всего было заказано

```python
pd.read_sql(
    """
    SELECT SUM(Quantity*TRIM(SUBSTR(Unit, 1, 3))) AS Tins  
    FROM Products AS pr INNER JOIN  
        OrderDetails AS ordt ON pr.ProductID = ordt.ProductID  
    WHERE ProductName LIKE "%Crab%" AND ProductName LIKE "%Meat%"  
    GROUP BY ProductName, Unit
    """, con)
```

добавлено наименование товара, количество единиц товара (коробок) и количество банок в единице товара

```python
pd.read_sql(
    """
    SELECT ProductName, SUM(Quantity) AS Boxеs, Unit, SUM(Quantity*TRIM(SUBSTR(Unit, 1, 3))) AS Tins  
    FROM Products AS pr INNER JOIN  
        OrderDetails AS ordt ON pr.ProductID = ordt.ProductID  
    WHERE ProductName LIKE "%Crab%" AND ProductName LIKE "%Meat%"  
    GROUP BY ProductName, Unit
    """, con)
```
