## Задание семинара: реализовать tfidf для quora.txt; найти самые схожие объекты (косинусное расстояние)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
#from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = list(open("quora.txt", encoding="utf-8"))
text = data[:100]
text[0]
```

```python
#using the count vectorizer
count = CountVectorizer()
word_count=count.fit_transform(text)
print(word_count)
```

```python
print(word_count.toarray())
```

```python
# считаем IDF
tfidf_transformer=TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer.fit(word_count)
df_idf = pd.DataFrame(tfidf_transformer.idf_, index=count.get_feature_names_out(),columns=["idf_weights"])
#inverse document frequency
df_idf.sort_values(by=['idf_weights'])
```

```python
#tfidf
tf_idf_vector=tfidf_transformer.transform(word_count)
feature_names = count.get_feature_names_out()
first_document_vector=tf_idf_vector[0]
matrix = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=[0])
for i in range(1, len(text)):
    tf_idf_vector=tfidf_transformer.transform(word_count)
    feature_names = count.get_feature_names_out()
    first_document_vector=tf_idf_vector[i]
    df_tfifd= pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=[i])
    matrix = pd.merge(matrix, df_tfifd, left_index=True, right_index=True)
# вывод tfidf для 50 предложений (строки - предложения, столбцы - фичи)
matrix_tfidf = matrix.T
matrix_tfidf
```

```python
from numpy import dot
from numpy.linalg import norm
cos = []
for i in range(len(matrix_tfidf)):
    string = []
    a = list(matrix_tfidf.iloc[i,:].values)
    for j in range(len(matrix_tfidf)):
        b = list(matrix_tfidf.iloc[j,:].values)
        string.append(dot(a, b)/(norm(a)*norm(b)))
    cos.append(string)
cos
```

```python
# матрица косинусных мер близости первых 50 предложений датасета quora
df = pd.DataFrame(cos)
df
```

```python
df = pd.DataFrame(cos)
for i in range(df.shape[0]):
    df.iat[i,i] = 0
str = df.max().idxmax()
col = df.loc[str,:].idxmax()
print("Наиболее схожие строки:")
print(text[str])
print(text[col])
print(f"Косинусная мера между строками: {df.max().max()}")
```

```python
def string_to_tfidf(phrase):
    # токенизация
    phrase_tok = tokenizer.tokenize(phrase.lower())
    # препроцессинг фразы
    phrase_lem_stem = []
    for w in phrase_tok:
        if w not in stop_words and w not in special_char:
            # лемматизация и стемминг русских слов
            word_lem_stem_rus = snowball.stem(morph.normal_forms(w)[0])
            # лемматизация и стемминг английских слов
            word_lem_stem_rus_eng = ps.stem((lemmatizer.lemmatize(word_lem_stem_rus)))
            phrase_lem_stem.append(word_lem_stem_rus_eng)
            string = (' '.join(phrase_lem_stem))
    phrase_preprocessed = [string]
    #using the count vectorizer
    phrase_words_count = count.transform(phrase_preprocessed) 
    #tfidf
    phrase_words_tfidf_short = tfidf_transformer.transform(phrase_words_count)
    phrase_words_tfidf = pd.DataFrame(phrase_words_tfidf_short.T.todense(), index=feature_names, columns=[0])
    return phrase_words_tfidf.T
```

```python
def predict_logreg(phrase):
    phrase_tfidf = string_to_tfidf(phrase) 
    if logreg.predict(phrase_tfidf) > 0.5:
        is_toxic = "The comment is toxic"
    else:
        is_toxic = "The comment isn't toxic"
    return is_toxic
```
